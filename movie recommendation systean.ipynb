```python
#Content based and popularity based recommendation systeam
# impoerting Dependencies
import pandas as pd 
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#data collection and preprocessing
# loading the data from the csv into pandas ddataframe_columns

movies_data = pd.read_csv(r'C:/Users/Admin/OneDrive/Desktop/pandas/movies.csv')
movies_data.head()
```

```python
#number of rows and columns in the ddataframe 
movies_data.shape
```

```python
# selectred the releavant features for recommendation
selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
print(selected_features)
```

```python
#replacing the null values with null string 

for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
#combined all the 5 selected features 
combined_feature = movies_data['genres']+''+movies_data['keywords']+''+movies_data['tagline']+''+movies_data['cast']+''+movies_data['director']
```

```python
print(combined_feature)
```

```python
#converting the text data into feature vector
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_feature)
print(feature_vectors)
```

```python
#cosine simalarity
# getting the simalarity score using cosine_similarity
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
similarity.shape
```

```python
#getting the mvoie name from the user 
movie_name = input('enter your favourate movie name :')
```

```python
#creating a list of all the movie names with the dataset 
list_of_movies = movies_data['title'].tolist()
print(list_of_movies)
```

```python
#finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(movie_name, list_of_movies)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
#finding the index of movie with title 
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting a list of similar movies   
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score, key= lambda x:x [1], reverse=True)
# print(sorted_similar_movies)
```

```python
# print the name of similar movies based on index  

print('Movie suggested for you  : \n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if (i<20):
          print(i, '.', title_from_index)
          i+=1
  
```

```python
# Movie Recomendation system 
movie_name = input('enter your favourate movie name :')

list_of_movies = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_movies)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key= lambda x:x [1], reverse=True)

print('Movie suggested for you  : \n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if (i<20):
          print(i, '.', title_from_index)
          i+=1
  
```
